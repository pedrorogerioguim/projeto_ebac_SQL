## **1.**Sobre a base de dados

Os dados contém informações sobre os clientes de um banco, organizados nas seguintes colunas:

*   **idade:** idade do cliente;
*   **sexo:**  gênero do cliente (feminino ou masculino);
*   **dependentes:** quantidade de filhos;
*   **escolaridade:** nível de escolaridade;
*   **salario_anual:** faixa salarial anual do cliente;
*   **tipo_cartao:** modalidade de cartão do cliente;
*   **qtd_produtos:** quantidade de produtos comprados nos últimos 12 meses;
*   **iteracoes_12m:** quantidade de transações nos últimos 12 meses;
*   **meses_inativo_12m:** quantidade de meses que o cliente ficou inativo;
*   **limite_credito:** limite de crédito fornecido ao cliente;
*   **valor_transacoes_12m:** valor total das transações realizadas nos últimos 12 meses;
*   **qtd_transacoes_12m:** quantidade de transações realizadas nos últimos 12 meses.

image.png

A maioria dos clientes deste banco (701 indivíduos) está situada na faixa salarial mais baixa, com ganhos anuais inferiores a 40 mil. Por outro lado, a menor quantidade de clientes (222 pessoas) tem renda anual superior a 120 mil. Há 235 clientes cuja faixa salarial não foi declarada ou apresenta dados nulos. ##**4.** Conclusão

Predomínio de Menores Rendimentos: A maioria dos clientes do banco possui uma renda anual mais baixa, especialmente na faixa de até 40K, indicando que o banco atende predominantemente a esse segmento econômico.

Diversidade Etária: Não há uma faixa etária predominante entre os clientes, sugerindo uma diversidade significativa na base de clientes.

Influência Limitada da Escolaridade: Parece não haver uma influência direta da escolaridade no limite de crédito, indicando que outros fatores podem estar mais fortemente associados a essa variável.

Gênero e Limites de Crédito: Há uma tendência de homens receberem maiores limites de crédito em comparação com as mulheres. Além disso, mulheres parecem ser mais representadas entre os clientes com menores limites de crédito, destacando uma disparidade de gênero.

Impacto da Faixa Salarial: A faixa salarial dos clientes parece ser um fator crucial na determinação do limite de crédito, indicando uma correlação direta entre renda e crédito oferecido.

Disparidade de Gênero e Salário: A observação de que não existem mulheres com salários acima de 60K que também não possuam os maiores limites de crédito levanta a questão de possíveis disparidades salariais de gênero que podem estar influenciando diretamente nos limites de crédito concedidos.

Essas conclusões apontam para possíveis áreas de atenção em termos de igualdade de gênero, representatividade e políticas de crédito, destacando a importância de avaliar e abordar possíveis disparidades no tratamento entre diferentes grupos de clientes.
